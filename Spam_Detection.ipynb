```python
import pandas as pd
df = pd.read_csv("spam.csv")
df.head()
```

```python
df.groupby('Category').describe()
```

```python
df['spam']=df['Category'].apply(lambda x: 1 if x=='spam' else 0)
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.Message,df.spam)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:2]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
emails = [
    'Hey mohan, can we get together to watch the football game tomorrow?',
    'Up to 20% discount on parking, exclusive offer just for you. Don\'t miss this reward!',
    'Congratulations, you have won a free trip to Hawaii! Click here to claim your prize.',
    'Important meeting tomorrow at 10 AM. Please be prepared with your presentation.'
]


# Transform the sample emails
emails_count = v.transform(emails)

# Predict whether the sample emails are spam or not
predictions = model.predict(emails_count)

# Print the predictions
for i, email in enumerate(emails):
    print(f"Email: {email}")
    if predictions[i] == 1:
        print("Predicted as spam")
    else:
        print("Predicted as not spam")
    print()
```

```python
X_test_count = v.transform(X_test)
model.score(X_test_count, y_test)
```

```python
from sklearn.pipeline import Pipeline
#uSE PIPELINE  INSTEAD OF COUNT VECTORZED METHOD
pipeline = Pipeline([
    ('vectorizer', CountVectorizer()),  # Vectorize the text data
    ('classifier', MultinomialNB())     # Use Multinomial Naive Bayes for classification
])
```

```python
MODEL = pipeline.fit(X_train, y_train)
MODEL.score(X_test, y_test)
```
