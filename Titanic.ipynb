```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
print(df.head())
```

```python
df.columns
```

```python
columns_to_drop = ['PassengerId', 'Name', 'Ticket', 'Cabin', 'Embarked', 'Parch', 'SibSp']
df = df.drop(columns_to_drop, axis=1)
print(df.head())
```

```python
input = df.drop(['Survived'], axis=1)
output = df['Survived']
```

```python
input.head()
```

```python
output.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
input['Sex'] = le.fit_transform(input['Sex'])
input.head()
```

```python
input.Age[:10]
```

```python
input.Age = input.Age.fillna(input.Age.mean())
input.Age[:10]
```

```python
input.isnull().sum()
input[1:10]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(input, output, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
accuracy = dt.score(X_test, y_test)
print("Accuracy:", accuracy)
```

```python
dt.predict(X_test)
```

```python
# Predict survival for the input data
input_data = [[1, 0, 38.0, 71.2833]]
predicted_value = dt.predict(input_data)[0]

# Check the predicted value
if predicted_value == 0:
    print("Predicted: Did not survive")
else:
    print("Predicted: Survived")
```
